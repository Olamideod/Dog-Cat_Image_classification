# **Split Dataset**

```python
train_data_dir =image_data
batch_size = 32
target_size = (224,224)
validation_split = 0.2
train= tf.keras.preprocessing.image_dataset_from_directory(
    train_data_dir,
    validation_split=validation_split,
    subset="training",
    seed=100,
    image_size=target_size,
    batch_size=batch_size,
)
validation= tf.keras.preprocessing.image_dataset_from_directory(
    train_data_dir,
    validation_split=validation_split,
    subset="validation",
    seed=200,
    image_size=target_size,
    batch_size=batch_size,
)
```

```python
class_names = train.class_names
class_names
```

```python
plt.figure(figsize=(15, 20))
for images, labels in train.take(1):
    for i in range(8):
        ax = plt.subplot(8, 4, i + 1)
        plt.imshow(images[i].numpy().astype("uint8"))
        plt.title(class_names[labels[i]])
        plt.axis("off")
```

**Modeling**

```python
base_model = tf.keras.applications.efficientnet_v2.EfficientNetV2B1(input_shape=(224,224,3),include_top=False,weights='imagenet')
base_model.trainable = False
keras_model=keras.models.Sequential()
keras_model.add(base_model)
keras_model.add(keras.layers.Flatten())
keras_model.add(keras.layers.Dropout(0.5))
keras_model.add(keras.layers.Dense(2,activation=tf.nn.softmax))
keras_model.summary()
```

```python
tf.keras.utils.plot_model(keras_model, to_file='model.png', show_shapes=True, show_layer_names=True,show_dtype=True,dpi=80)
```

```python
checkpoint =ModelCheckpoint("my_keras_model.h5", save_best_only=True)
```

```python
early_stopping =EarlyStopping(patience=5, restore_best_weights=True)
```

```python
keras_model.compile(optimizer ='Adam',loss='sparse_categorical_crossentropy', metrics=['accuracy'])
hist=keras_model.fit_generator(train,epochs=40,validation_data=validation,callbacks=[checkpoint,early_stopping])
```

```python
hist_=pd.DataFrame(hist.history)
hist_
```

```python
score, acc = keras_model.evaluate(validation)
print('Test Loss =', score)
print('Test Accuracy =', acc)
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.image import load_img, img_to_array
import numpy as np

# Load the pre-trained VGG16 model
model = keras.applications.VGG16(weights="imagenet")

# Specify the correct file path for the uploaded image
image_path = "download.jpeg"

# Load and preprocess the uploaded image
image = load_img(image_path, target_size=(224, 224))
image = img_to_array(image)
image = np.expand_dims(image, axis=0)
image = keras.applications.vgg16.preprocess_input(image)

# Make predictions
predictions = model.predict(image)
decoded_predictions = keras.applications.vgg16.decode_predictions(predictions, top=5)[0]

# Display the image
import matplotlib.pyplot as plt
plt.imshow(load_img(image_path))
plt.axis('off')
plt.show()

# Display the top predicted labels
for label, description, score in decoded_predictions:
    print(f"{description}: {score:.2f}")
```
